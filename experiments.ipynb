```python
from sklearn.metrics import RocCurveDisplay, confusion_matrix
import joblib, numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
model = joblib.load("model/model.joblib")
```

```python
data = np.load("pred_data/data.npz")
xtrain, xtest , ytrain, ytest, preds = data['xtr'], data['xts'], data['ytr'], data['yts'], data['preds']
```

```python
RocCurveDisplay.from_estimator(model, xtest, ytest)
```

```python
confusionmatrix_ = confusion_matrix(ytest, preds)
fig, ax = plt.subplots(figsize = (3,3))
ax = sns.heatmap(confusionmatrix_, annot=True, cbar = False)
plt.xlabel("predictions")
plt.ylabel("true labels");
```
